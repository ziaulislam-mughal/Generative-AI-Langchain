```python
from langchain_community.tools import tool
```

### 1. Add Doc string 

```python
def addition(a,b):
    "this function is used to find addition of two numbers."
    return a+b
```

### 2. Add Type Hint

```python
def addition(a:int , b:int) -> int:
    "this function is used to find addition of two numbers."
    return a+b
```

```python
@tool
def addition(a:int , b:int) -> int:
    "this function is used to find addition of two numbers."
    return a+b
```

```python
result = addition.invoke({'a':4 , 'b':5})
print(result)
```

```python
print(addition.args)
print(addition.description)
print(addition.name)
```

```python
print(addition.args_schema.model_json_schema())
```

### 2. Structure Tool 

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class multiplyinput(BaseModel):
    a : int = Field(..., Description = "Enter a for addition.")
    b : int = Field(..., Description = "Enter b for addition.")
```

```python
def multiply(a:int,b:int) -> int:
    "this function is used to muliply these numbers"
    return a * b
```

```python
from langchain_core.tools import tool
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply,
    name="multiply",
    description="Multiply two numbers",
    args_schema=multiplyinput
)
```

```python
result = multiply_tool.invoke({'a':5 , 'b':4})
print(result)
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

4. BaseTool Model 

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
class Multinput(BaseModel):
    a : int = Field(required = True , description="this is number used to multiply with others number.")
    b : int = Field(required = True , description="this is number used to multiply with others number.")
```

```python
class MultiplyTool(BaseTool):
    name : str = "multiply"
    description : str = "Multiply two numbers"
    args_schema : Type[BaseModel] = Multinput

    
```

```python
def _run(self, a : int , b : int )-> int:
    return a * b 
```

```python
multiply_tool = MultiplyTool()
```
