```python
from langchain_community.tools import tool
```

### 1. Add Doc string 

```python
def addition(a,b):
    "this function is used to find addition of two numbers."
    return a+b
```

### 2. Add Type Hint

```python
def addition(a:int , b:int) -> int:
    "this function is used to find addition of two numbers."
    return a+b
```

```python
@tool
def addition(a:int , b:int) -> int:
    "this function is used to find addition of two numbers."
    return a+b
```

```python
result = addition.invoke({'a':4 , 'b':5})
print(result)
```
