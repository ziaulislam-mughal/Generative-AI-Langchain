```python
import os
from dotenv import load_dotenv
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.messages import HumanMessage
from langchain_core.tools import tool
import requests
```

```python
load_dotenv()

# 2. Check Token
token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
if not token:
    print("Error: HUGGINGFACEHUB_API_TOKEN is not set in the .env file.")

# 3. Setup the Endpoint
repo_id = "CohereForAI/c4ai-command-r-plus"
llm = HuggingFaceEndpoint(
    repo_id=repo_id,
    task="text-generation",
    max_new_tokens=512,
    temperature=0.7,
    huggingfacehub_api_token=token,
)

# 4. Wrap it in ChatHuggingFace (This handles the "conversational" format for you)
model = ChatHuggingFace(llm=llm)
```

```python
print("Token loaded:", token[:10])
```

Create Tool

```python
@tool
def addition(a:int , b:int) -> int:
    "this function is used to add two numbers a and b. "
    return  a+b 

@tool
def multiplication(a:int , b:int ) -> int :
    "this function is used to multiply two numbers a and b"
    return a * b 
```

```python
multiplication.invoke({'a': 5 ,
                       "b":5})
```

```python
multiplication.args
```

bind

```python
llm_with_tools = model.bind_tools([addition , multiplication])
```

call

excution 

```python
from langchain_core.messages import HumanMessage

query = HumanMessage(content="can i calculate addition 4 with 10 ?")

messages = [query]

result = llm_with_tools.invoke(messages)

print(result)
```
